```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df = pd.read_csv("dataset/diabetes.csv")
df.dropna(inplace=True)
df.head(10)
```

```python
df.info()
```

```python
df.describe().T
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

print("Panjang X = ", len(X))
print("Panjang y = ", len(y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print("Jumlah data X_train = ", len(X_train))
print("Jumlah data X_test = ", len(X_test))
```

```python
k_range = range(1, 21)
scores_list = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores_list.append(metrics.accuracy_score(y_test, y_pred))

scores_data = {"k": k_range, "score": scores_list}
scores_pd = pd.DataFrame(scores_data)
print(scores_pd)
```

```python
max_temp = max(scores_list)
for k in enumerate(scores_list, 1):
    if k[1] == max_temp:
        max_score = (k[0], k[1])
        break

print("Akurasi maksimal menggunakan K-Nearest Neighbors adalah {:.2f}".format(max_score[1]*100), "% dengan jumlah k (tetangga terdekat) adalah", max_score[0])
```

```python
knn = KNeighborsClassifier(n_neighbors=max_score[0])
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=["Predicted"], margins=True)
```
